```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
z=[10,20,30,40]
c=["r",'y','g','b']
plt.xlabel("LANGUAGE",)
plt.ylabel("no")
plt.title("ankit yadav",fontsize=40)
plt.bar(x,y,color="g",width=0.4,align='center',edgecolor="r",linewidth=5,linestyle=":",alpha=0.4
       ,label='ankit')

plt.bar(x,z,color="r",label='company')
plt.legend()
plt.show()
```
