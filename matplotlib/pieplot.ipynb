```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=['c','c++','js','python']

plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=['c','c++','js','python']
plt.pie(x,labels=y)
plt.show()
```
