```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[5,8,3,9,2,8,1]
plt.scatter(x,y)
plt.show
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[5,8,3,9,2,8,1]
plt.scatter(day,no,color='b')
plt.title('scatter plot',fontsize=15)
plt.xlabel("day")
plt.show
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[5,8,3,9,2,8,1]
colors=[10,54,92,41,45,75,96]
plt.scatter(day,no,c=colors,cmap='BrBG')
plt.colorbar()
plt.title('scatter plot',fontsize=15)
plt.xlabel("day")
plt.show
```
