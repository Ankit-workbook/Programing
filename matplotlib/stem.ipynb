```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,3]

plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,3]

plt.stem(x,y,linefmt=':')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,3]

plt.stem(x,y,markerfmt='r+')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,3]

plt.stem(x,y,bottom=9)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,3]

plt.stem(x,y,orientation="horizontal")
plt.show()
```
