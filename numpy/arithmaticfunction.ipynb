```python
import numpy as np 

var = np.array([13  ,8 ,10, 13])

print("min:",np.min(var))
print("max:",np.max(var))
```

```python
var = np.array([13  ,8 ,10, 13])

print("min:",np.min(var),np.argmin(var))
print("max:",np.max(var),np.argmax(var))
```

```python
var11= np.array([[1,2,3,4],[1,2,3,4]])
print(np.min(var11,axis=1))
```
