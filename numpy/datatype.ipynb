```python
import numpy as np 


var2 = np.array([13  ,8 ,10, 13])
print("data type:",var2.dtype)
```

```python

var2 = np.array([13  ,8 ,10, 1.3])
print("data type:",var2.dtype)
```

```python

var2 = np.array([13  ,'A',10, 13])
print("data type:",var2.dtype)
```

```python
var2 = np.array([13 ,10, 13])
print("data type:",var2.dtype)
```

```python
var2 = np.array([13 ,10, 13],dtype=np.int8)
print("data type:",var2.dtype)
print(var2)
```

```python
var2 = np.array([13 ,10, 13],dtype="f")
print("data type:",var2.dtype)
print(var2)
```

```python
x= np.array([1,2,3,4])
new=np.float32(x)

print("data type:",x.dtype)
print("data type:",new.dtype)
print(x)
print(new)
```

```python
x= np.array([1,2,3,4])
new=np.float32(x)
new1=np.int_(new)

print("data type:",x.dtype)
print("data type:",new.dtype)
print("data type:",new1.dtype)
print(x)
print(new)
print(new1)
```
